```python
# TASK 2
import requests
import shutil
import os

accidents_folder = 'accident_all_years'
os.mkdir(accidents_folder)

base_url = 'https://static.nhtsa.gov/nhtsa/downloads/FARS/{year}/National/FARS{year}NationalCSV.zip'

for year in range(1975, 2021, 5):

    url = base_url.format(year=year)
    r = requests.get(url)
  
    filename = url.split('/')[-1]
    
    with open(filename, 'wb') as output_file:
        output_file.write(r.content)
    print(f'downloaded {filename}')

    shutil.unpack_archive(filename, str(year))
    print(f'unzipped {filename} into folder {year}')

    try:
        shutil.copyfile(f'{year}/ACCIDENT.CSV', f'{accidents_folder}/{year}_accident.csv')
        print(f'moved file {year}/ACCIDENT.CSV to {accidents_folder}/{year}_accident.csv')
    except:
        try:
            shutil.copyfile(f'{year}/accident.csv', f'{accidents_folder}/{year}_accident.csv')
            print(f'moved file {year}/accident.csv to {accidents_folder}/{year}_accident.csv')
        except:
            shutil.copyfile(f'{year}/accident.CSV', f'{accidents_folder}/{year}_accident.csv')
            print(f'moved file {year}/accident.CSV to {accidents_folder}/{year}_accident.csv')   
```

```python
import pandas as pd
import glob
import os

# Adding full year
for year in range(1975, 2021, 5):
    df = pd.read_csv(f'accident_all_years/{year}_accident.csv', encoding="cp1252")
    df['YEAR'] = year
    df.to_csv(f'accident_all_years/{year}_accident.csv', index=None, encoding='utf-8')

# merging the files
joined_files = os.path.join("accident_all_years", "*_accident.csv")

# A list of all joined files is returned
joined_list = glob.glob(joined_files)

# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True) 
```

```python
# You will create a subset of the data over all years.
# • include 5000 random rows of data
# • restrict data only to the following columns:
# 'STATE',
# 'MONTH',
# 'DAY',
# 'YEAR',
# 'HOUR',
# 'PERSONS',
# 'MAN_COLL',
# 'LGT_COND',
# 'WEATHER',
# 'SCH_BUS',
# 'FATALS',
# 'DAY_WEEK',
# 'DRUNK_DR',
# • you may need to eliminate rows with NaN data to simplify the analysis

df_5000 = df.sample(5000)

df_subset = df_5000[['STATE',
'MONTH',
'DAY',
'YEAR',
'HOUR',
'PERSONS',
'MAN_COLL',
'LGT_COND',
'WEATHER',
'SCH_BUS',
'FATALS',
'DAY_WEEK',
'DRUNK_DR',
]]

df_subset.dropna(inplace=True)
```

```python
# Scale the data from part (a) such that all values are between 0 and 1
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(df_subset)
scaled_subset = scaler.transform(df_subset)

scaled_subset
```

```python
# How many components are necessary to capture 90% of the data in the datast from the prior section. Show the plot of
# the cummulative variance using PCA.explained_variance_ratio_
from sklearn.decomposition import PCA

N = list(range(1, 14))

pca = PCA(n_components=13)
X = pca.fit_transform(scaled_subset)
explained_variances=pca.explained_variance_ratio_

df_pca = pd.DataFrame({'N': N, 'explained_variances': explained_variances})
df_pca.plot('N', 'explained_variances', kind='line')
```

```python
import plotly.express as px
from sklearn.decomposition import PCA

features = df_subset.columns

pca = PCA()
components = pca.fit_transform(df_subset[features])
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    labels=labels,
    dimensions=range(len(features)),
    color=df_subset['STATE']
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
